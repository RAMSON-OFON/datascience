Because we need to perform web scraping, we will need to import the libraries shown below. The urllib.request module is used to open URLs. The Beautiful Soup package is used to extract data from html files. The Beautiful Soup library's name is bs4 which stands for Beautiful Soup, version

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
site = urlopen(url)
```

```python
soup = BeautifulSoup(site, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
#if we are interested in the text of the page,
text = soup.get_text()
#print(text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
table_rows = soup.find_all('tr')
print(table_rows[:10])
```

```python
for row in table_rows:
    td = row.find_all('td')
print(td)

print(type(td))
```

```python
str_cells = str(td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in table_rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
df1[0] = df1[0].str.strip(']')
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df6 = df5.dropna(axis=0, how='any')
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename (columns ={'[Place': 'Place'}, inplace = True)
```

```python
df7.rename(columns = {'Team]': 'Team'}, inplace = True)
```

The first question to answer is, what was the average finish time (in minutes) for the runners? You need to convert the column "Chip Time" into just minutes. One way to do this is to convert the column to a list first for manipulation.

```python
time_list = df7[' Chip Time'].tolist()

# You can use a for loop to convert 'Chip Time' to minutes

time_mins = []
for i in time_list:
    h, m, s = i.split(':')
    math = (int(h) * 3600 + int(m) * 60 + int(s))/60
    time_mins.append(math)
#print(time_mins)
```
